```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Downloads/Iris.csv")
df.head()
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
#use label encoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
ss = le.fit_transform(df["Species"])
ss
```

```python
df['Species_lable']  = le.fit_transform(df["Species"])
df
```

```python

X1 = df.drop('Species',axis=1)
X = X1.drop('Species_lable',axis=1)
y = X1['Species_lable']
X1.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X1,y,test_size=0.2,random_state=51)
```

```python
plt.scatter(X.loc[:,"SepalLengthCm"], y , color='red')
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
X_train[:,0].shape
type(y_train)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg =  PolynomialFeatures(degree=2)
X_train_poly = poly_reg.fit_transform(X_train)
X_test_poly = poly_reg.transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_poly,y_train)
```

```python
lr.predict(X_test)
```

```python
afd = pd.DataFrame(X_test,columns=['1','2','3','4','5'])
afd
```

```python
lr.score(X_test_poly,y_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X_test_poly[:,1], y_test , color='red')
plt.plot(X_train,y_train,'o')
plt.plot(X_train,y_train)

plt.show()
```
